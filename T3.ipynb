# Pregunta 6 ### 
Al hacer una tabla de frecuencia sobre la cantidad total de camas críticas en los hospitales de la red, considerando intervalos de 3 en 3, ¿en cuántos intervalos hay exactamente 1 cama crítica? 

```python
apoyo6 = Hosp_sheet[["Nombre Establecimiento", "Total Camas Críticas"]].groupby("Nombre Establecimiento").sum()
apoyo6.plot(kind='barh', legend = None, figsize = (15,5))
```

```python
import pandas as pd

# Supongamos que tienes un DataFrame llamado df con la columna 'Camas Críticas'

# Definir los límites de los intervalos
min_value = Hosp_sheet['Total Camas Críticas'].min()
max_value = Hosp_sheet['Total Camas Críticas'].max()
intervalos = range(min_value, max_value + 3, 3)

# Agrupar los valores en intervalos y contar la frecuencia
tabla_frecuencias = pd.cut(Hosp_sheet['Total Camas Críticas'], bins=intervalos, include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns = ['Intervalo', 'Frecuencia']

# Contar el número de intervalos con exactamente 1 cama crítica
numero_intervalos_una_cama = tabla_frecuencias[tabla_frecuencias['Frecuencia'] == 1].shape[0]

# Mostrar el número de intervalos con exactamente 1 cama crítica
print("Número de intervalos con exactamente 1 cama crítica:", numero_intervalos_una_cama)
tabla_frecuencias
```
