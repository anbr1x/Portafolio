```python
#Seperating unique errors.
errors = set()
for i in tr_l['errors'].unique().tolist():
    for j in i.split(','):
        errors.add(j)
errors = list(errors)
errors
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
# Divide los errores en listas
tr_l['errors_split'] = tr_l['errors'].str.split(',')
# Aplica el MultiLabelBinarizer de manera vectorizada
errors_transformed = mlb.fit_transform(tr_l['errors_split'])
# Crea un DataFrame con los resultados transformados
errors_df = pd.DataFrame(errors_transformed, columns=mlb.classes_)
# Concatena el DataFrame original con el DataFrame de errores transformados
tr_error_lb = pd.concat([tr_l, errors_df], axis=1)
# Elimina la columna temporal 'errors_split'
tr_error_lb.drop(columns=['errors_split'], inplace=True)
```

### Preprocessing Date & Time


```python
dt_tr = pd.merge(tr_error_lb[['id', 'client_id', 'card_id', 'date']], cards_l[['card_id', 'expires', 'acct_open_date', 'year_pin_last_changed']], on='card_id', how='inner')
dt_tr.head()
```

```python
dt_tr['expires'] = pd.to_datetime(dt_tr['expires'], format='%m/%Y')
dt_tr['acct_open_date'] = pd.to_datetime(dt_tr['acct_open_date'], format='%m/%Y')
dt_tr['year_pin_last_changed'] = pd.to_datetime(dt_tr['year_pin_last_changed'], format='%Y')
```

### Feature Enginnering

```python
## days para expirar de la tarjeta en el momento de la transaccion
dt_tr['days_to_expire'] = ((dt_tr['expires'].dt.year - dt_tr['date'].dt.year) * 12) + (dt_tr['expires'].dt.month - dt_tr['date'].dt.month) #(in months)
## antiguedad de la tarjeta al momento de la transaccion
dt_tr['opened_since'] = ((dt_tr['date'].dt.year - dt_tr['acct_open_date'].dt.year) * 12) + (dt_tr['date'].dt.month - dt_tr['acct_open_date'].dt.month) #(in months)
## ant en meses desde la ultima actualizacion de pin
dt_tr['pin_updated_since'] = (dt_tr['year_pin_last_changed'].dt.year - dt_tr['date'].dt.year) * 12 #(in months.)
```

```python
dt_tr['is_expired'] = np.where(dt_tr['days_to_expire'] < 0,1,0)
```

```python
dt_tr['tr_hour'] = dt_tr['date'].dt.hour
dt_tr['tr_weekday'] = dt_tr['date'].dt.weekday
dt_tr['tr_month'] = dt_tr['date'].dt.month
```

```python
dt_tr.drop(['date', 'expires', 'acct_open_date', 'year_pin_last_changed', 'client_id', 'card_id'], axis=1, inplace=True)
```

```python
# Funcion que genera reporte del modelo
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, ConfusionMatrixDisplay, roc_auc_score, roc_curve
def generate_model_report(y_pred, model):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)

    print("Accuracy:-", accuracy)
    print("Precision:-", precision)
    print("Recall:-", recall)
    print("F1 Score:-", f1)
    print("ROC AUC Score:", roc_auc)

    cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
    disp = ConfusionMatrixDisplay(cm, display_labels=model.classes_)
    disp.plot()

    y_scores = model.predict_proba(x_test)[:, 1]

    fpr, tpr, thresholds = roc_curve(y_test, y_scores)
    roc_auc = roc_auc_score(y_test, y_scores)

    plt.figure(figsize=(8, 6))
    sns.set(style='whitegrid')
    sns.lineplot(x=fpr, y=tpr, label=f'XGBoost (AUC = {roc_auc:.4f})', color='darkorange', lw=2)
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')

    plt.title('ROC AUC Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate (Recall)')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.show()
```
