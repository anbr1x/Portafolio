# Pregunta1
#### ¿Cuál opción corresponde al servicio de salud al que llegaron más pacientes el día 1?  



```python
import pandas as pd
```

```python
# Leer una hoja específica del archivo Excel
Hosp_sheet = pd.read_excel('Datos.xls', sheet_name='Hospitales')
Hosp_sheet
```

```python
HospitalxServ = Hosp_sheet[["Nombre Servicio de Salud","Nombre Establecimiento"]].groupby("Nombre Servicio de Salud") 
#de aqui saco las columnas que necesito para luego agruparlas
HxS = HospitalxServ.apply(lambda x: x).reset_index(drop=True)
```

```python
Info_sheet = pd.read_excel('Datos.xls', sheet_name='Información Histórica') #Datos del dia de llegada
Info_sheet
```

```python
df1 = Info_sheet.iloc[:146] #Recorto solo los del primer dia
#agrupo y cuento
result = df1.groupby(['Día de llegada', 'Hospital al que llega']).size().reset_index(name='Num_Pacientes') 
result.sort_values(by=["Num_Pacientes"],ascending = False) #Ordeno de mayor a menor
```

```python
result = result.rename(columns={'Hospital al que llega' : 'Nombre Establecimiento'}) #primero modifico el nombre de las columnas para que se pueda hacer un merge
result
```

```python
df_combinado = pd.merge(result, HxS, on='Nombre Establecimiento') #combino ambos df por la columna comun
df_combinado #DF combinado
```

```python
result1 = df_combinado.groupby('Nombre Servicio de Salud')['Num_Pacientes'].sum().reset_index(name='Total_Pacientes')#Por ultimo agrupo y sumo los pacientes por servicio de salud
result1
```

# Pregunta7
### ¿Qué servicio de salud gastó más dinero, para la atención de los pacientes que llegaron los últimos 10 días utilizando los costos diarios de atención por GRD promedio de cada sistema (desde el día 415 al día 424)?  #### Se vuelve generica la 6


```python
df_combinado6
```

```python
result7 = df_combinado6.groupby(['Nombre Servicio de Salud','GRD',"Estadia"]).size().reset_index(name='Num_Pacientes')
result7 #para mergear con la columna de costos medios de GRD
```

```python
result7['Etiqueta'] = result7['GRD'] + '-' + result7['Nombre Servicio de Salud']
result7_1 = result7.iloc[:,2:-1]
nombres_columnas = ['Estadia', 'Num_Pacientes','Etiqueta']
result7_1.columns = nombres_columnas
result7_1
```

```python
GRD7 = Hosp_sheet.groupby('Nombre Servicio de Salud')[GRDs].mean()
GRD7.transpose()
GRD7_index = GRD7.transpose().reset_index(drop=False)
nuevos_nombres = ['GRD'] + list(GRD7_index.columns[1:])
GRD7_index.columns = nuevos_nombres
GRD7_index
```

```python
GRD7_index["GRD"] = ["GRD1","GRD2","GRD3", "GRD4","GRD5","GRD6","GRD7","GRD8","GRD9","GRD10"]
GRD7_index
```

```python
GRD7_index
```

```python
# Obtener las etiquetas de las columnas (excepto la primera)
columnas = GRD7_index.columns[1:]
# Crear una lista para almacenar los datos del DataFrame resultante
datos_resultantes = []

# Iterar sobre las columnas y concatenar la primera columna original con el nombre de cada columna
for columna in columnas:
    etiqueta = GRD7_index['GRD'].astype(str) + '-' + columna
    valores = GRD7_index[columna].tolist()
    datos_resultantes.extend(list(zip(etiqueta, valores)))

# Crear el DataFrame resultante
df_resultante = pd.DataFrame(datos_resultantes, columns=['Etiqueta', 'Valor'])

# Mostrar el DataFrame resultante de 60x2
df_resultante
```

```python
Reult7_F = pd.merge(df_resultante, result7_1, on='Etiqueta') #combino ambos df por la columna comun
Reult7_F['Costo'] = Reult7_F['Valor'] * Reult7_F['Estadia'] * Reult7_F['Num_Pacientes']
Reult7_F
```

```python

# Definir el tamaño del intervalo
intervalo = 10

# Calcular la suma en intervalos
suma_intervalos = []
for i in range(0, len(Reult7_F), intervalo):
    suma = Reult7_F.loc[i:i+intervalo-1, 'Costo'].sum()
    suma_intervalos.append(suma)

# Mostrar los resultados
suma_intervalos
```
