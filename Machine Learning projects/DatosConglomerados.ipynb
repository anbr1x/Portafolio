```python
import pandas as pd
from sklearn.utils import resample
```

```python
df0 = pd.read_excel('Archivos/EAUDP 17.10.23.xlsx')
df1= pd.read_excel('Archivos/EAUPDATE14.11.xlsx')
df2 = pd.read_excel('Archivos/PEA al 08.11.2023.xlsx',sheet_name='DATA1')
df3 = pd.read_excel('Archivos/Pendiente de entrada al 23.10.2023.xlsx',sheet_name='DATA')
```

```python
def Filtro(df):
    df_e = df.loc[df['SOLICITANTE'] == 'Eduardo Iberico']
    df_f = df_e[['CONCATENADO','TEXTO','CE_GESTOR','TIPO_PROYECTOS','NOMBRE PROYECTO','CLASIF_RED_1','CLASIF_RED_2','RESPONSABLE EA']]
    df_c = df_f.dropna(subset=['RESPONSABLE EA'])
    return df_c
```

```python
Data = [df0,df1,df2,df3]
Data_f= []
```

```python
for df in Data:
    df1 = Filtro(df)
    Data_f.append(df1)
    
```

```python
combined_df = pd.concat([Data_f[0], Data_f[1], Data_f[2], Data_f[3]])
```

```python
combined_df = combined_df.drop_duplicates()
```

```python
combined_df['RESPONSABLE EA'] = combined_df['RESPONSABLE EA'].str.upper()
```
