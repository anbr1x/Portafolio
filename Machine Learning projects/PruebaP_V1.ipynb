```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Supongamos que 'df' es tu DataFrame y que 'nombre_proyecto', 'descripcion' y 'responsable' son las columnas
df = pd.read_csv('4Predict_Analista.csv')
```

```python
df1 = df.drop('RESPONSABLE EA',axis=1)
```

```python
df1.info()
```

```python
# Convertir las columnas categóricas a numéricas
enc = OneHotEncoder()
categorical_features = df1.columns
transformed_data = enc.fit_transform(df[categorical_features])
feature_names = enc.get_feature_names_out(input_features=categorical_features)  # Asegúrate de pasar los mismos nombres de características utilizados en fit_transform
enc_df = pd.DataFrame(transformed_data.toarray(), columns=feature_names)
                                        
# Unir el DataFrame codificado con el original
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(enc_df, df['RESPONSABLE EA'], test_size=0.2, random_state=42)
```

```python

# Entrenar el modelo
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Hacer predicciones
y_pred = model.predict(X_test)

# Calcular la precisión
accuracy = accuracy_score(y_test, y_pred)
print(f'Precisión del modelo: {accuracy}')
```
